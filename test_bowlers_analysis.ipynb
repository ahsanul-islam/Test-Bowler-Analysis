# Interpretation of Each Column

Column no 0: Index
<br>
Column no 1: Player : Name of the player and country in brackets
<br>
Column no 2: Span   : Debut year to retirement year
<br>
Column no 3: Mat    : Number of matches the player played
<br>
Column no 4: Inns   : Number of innings the player played
<br>
Column no 5: Balls  : Number of balls the bowler delivered
<br>
Column no 6: Runs   : Total runs the bowler gave through his career
<br>
Column no 7: Wkts    : Total number of wickets the bowler took
<br>
Column no 8: BBI    : Best bowling figure in an innnings
<br>
Column no 9:BBM     : Best bowling figure in a match
<br>
Column no 10: Ave   : Average = (Runs/Wkts)
<br>
Column no 11: Econ  : Average run given in a single over by the bowler
<br>
Column no 12: SR    : Strike Rate=(Balls/Wickets)
<br>
Column no 13: 5     : Number of times where bowler had taken 5 wickets or more in a single innings
<br>
Column no 14: 10     : Number of times where bowler had taken 10 wickets or more in a single match
